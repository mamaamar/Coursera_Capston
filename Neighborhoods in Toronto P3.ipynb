## Importing the required librairies

```python
# import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# map rendering library
#!conda install -c conda-forge folium=0.5.0 --yes
import folium

#!pip install geopy

print('Libraries imported.')
```

## Importing the Postal Codes and Neighborhoud Data from Wikipedia

```python
pc_list = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
pc_data = pd.DataFrame(pc_list[0])
```

```python
pc_data.head()
```

## Removing the Borough with 'Not assigned' values
## Replacing the cell that has a borough but a 'Not assigned' neighborhood, then the neighborhood will be the same as the borough.

```python
pc_data = pc_data[pc_data['Borough'] != 'Not assigned']
pc_data['Neighborhood'].replace('Not assigned',pc_data['Borough'],inplace=True)
pc_data.head()
```

## Renaming the column Postal Code to PostalCode

```python
pc_data.rename(columns={"Postal Code": "PostalCode"}, inplace=True)
pc_data.head(10)
```

## Reinitializing the index to 0

```python
toronto_data = pc_data.reset_index(drop=True)
toronto_data.head()
```

## Printing the rows and columns of the DataFrame (12)

```python
toronto_data.head(12)
```

## Using the .shape method to print the number of rows of your dataframe.

```python
toronto_data.shape
```

## Getting the Goespastial data for Toronto neighborhouds

```python
neighborhood_coords = pd.read_csv('http://cocl.us/Geospatial_data')
neighborhood_coords.head()
```

## Renaming the column Postal Code to PostalCode

```python
neighborhood_coords.rename(columns={"Postal Code": "PostalCode"}, inplace=True)
neighborhood_coords.head(10)
```

## Merging Toronto neighborhouds data with geospatial data

```python
neighborhood_data = pd.merge(toronto_data, neighborhood_coords, on='PostalCode')
neighborhood_data.head()
```

## Displaying the number of boroughs and neighborhoods.

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(neighborhood_data['Borough'].unique()),
        neighborhood_data.shape[0]
    )
)
```

# Generating maps to visualize your neighborhoods and how they cluster together.
## Using Toronto geo coordinates (latitude = 43.651070, longitude= -79.347015)

```python
# create map of Toronto using latitude and longitude values
latitude = 43.651070
longitude= -79.347015

map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhood_data['Latitude'], neighborhood_data['Longitude'], neighborhood_data['Borough'], neighborhood_data['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
